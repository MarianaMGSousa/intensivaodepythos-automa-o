### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
!pip install pyautogui!pip install pyperclip
```

```python
import pyautogui
import pyperclip
import time   # permite pausar o código por alguns segundos, apenas naquela linha do código, naquele lugar específico


pyautogui.PAUSE = 1  # espera 1 segundo para executar cada linha - faz essa pausa em todo o código
# escreva o passo a passo em português
# passo 1 - Entrar no sistema da empresa (no caso o drive)
pyautogui.hotkey("ctrl", "t")  # hotkey para um atalho do teclado com o navegador aberto
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga") # para copiar um texto
pyautogui.hotkey("ctrl", "v")   # para colar o texto
pyautogui.press("enter") # pressiona enter no teclado - apenas um botão

time.sleep(3)


# passo 2 - Navegar no sistema até encontrar a base de dados
pyautogui.click(x=329, y=261, clicks=2)  # para saber a coordenada vamos la embaixo e descobre
time.sleep(2)    # como a base de dados pode demorar um pouco então colocamos essa pausa de 2 segundos

# passo 3 - Exportar a base de vendas
pyautogui.click(x=346, y=258)  # localização do arquivo
pyautogui.click(x=1165, y=159)   # Localização do 3 pontos, onde faz o download
pyautogui.click(x=951, y=570)  # Local para fazer dowmload
time.sleep(7)   # Como download pode demorar então vamos esperar esses 7 segundos
# Se quiser clicar com o botão direito do mouse, após a coordenada colocamos button="right"


```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Quantidade de Produtos



### Vamos agora enviar um e-mail pelo gmail

```python
# passo 4 - Calcular os indicadores (faturamento e quantidade de produtos vendidos)
import pandas as pd   #sempre que for usar base de dados precisa importar o pandas - esse pd é um apelido 

tabela = pd.read_excel(r"C:\Users\maria\Downloads\Vendas - Dez.xlsx") 
                       #sempre que for passar um endereço de um arquivo que esta no computador precisa colocar o r no início
                       # Se for outra aba da tabela coloca a virgula e depois sheets=3 (número ou nome da aba)
display(tabela) # para exibir a tabela                       

faturamento = tabela["Valor Final"].sum()   # usou [] pq foi uma lista
quantidade = tabela["Quantidade"].sum()
# display(faturamento)  - para aparecer o valor da soma
# display(quantidade)


# passo 5 - Enviar um email para a diretoria com os indicadores
# para entrar no email
pyautogui.hotkey("ctrl", "t")  # hotkey para um atalho do teclado com o navegador aberto
pyperclip.copy("https://mail.google.com/mail/u/0/?hl=pt-BR#inbox") # para copiar um texto
pyautogui.hotkey("ctrl", "v")   # para colar o texto
pyautogui.press("enter") # pressiona enter no teclado - apenas um botão
time.sleep(5)

#clicar no botão escrever
pyautogui.click(x=86, y=167)
time.sleep(3)

#escrever destinatário
pyautogui.write("marianamoreira2608@gmail.com")
pyautogui.press("tab")

#escrever assunto
pyautogui.press("tab")
pyperclip.copy("Relatório de vendas")
pyautogui.hotkey("ctrl", "v")

# escrever o corpo do email
pyautogui.press("tab")
# texto sempre começa e termina com 3 aspas
# f no início do texto serve para fornatar
# {} informa a variável
# , separador de milhar
# .2f é a quantidade de casas decimais
texto =f"""
Prezados, bom dia

O faturamento de ontem foi de R$ {faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Mariana Moreira
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar o email
pyautogui.hotkey("ctrl", "enter")

# Agora vamos rodar o código todo para garantir que esta tudo ok
# Vai em Cell + Run All

# Agora vamos programar para rodar diariamente, semanalmente, quinzenalmente, mensalmente
```
